```python
#It is a pre-defined library. It contains functions related to database and load full functions in memory and space consumption.
import pandas as pd
```

Data preprocessing

```python
#Assign the variable to read the excel csv file.It is called dataset/information into process - Data preprocessing
dataset=pd.read_csv("Salary_Data.csv")
#used to call the datas stored in variable.
dataset
```

Input and output splitting

Input split

```python
#It is Spliting the dataset of inputs and assign variable.
independent=dataset[["YearsExperience"]]
#used to call the input datas stored in variable.
independent
```

Output split

```python
#It is Spliting the dataset of outputs and assign variable.
dependent=dataset[["Salary"]]
#used to call the output datas stored in variable.
dependent
```

```python
Train and Test split
```

```python
#It is a library which supports for Machine learning-importing particular function which we need like (train_test_Split)& also 
#it is memory saver.
from sklearn.model_selection import train_test_split
#imported functions splitted and parameters are passed and assigning the variables as 4(X_train,X_test,Y_train,Y_test).
#need seperate variable to access next step
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

Model Creation

```python
#for model creation import linear regression fron library particular function. It is procedure and learning phase
from sklearn.linear_model import LinearRegression
#assign the variable(it is usual word in all company) for non-parametrised class function 
regressor=LinearRegression()
#Call the fit function and passes the parameters for subtution, weight and bias calculation done here.
regressor.fit(X_train,Y_train)
```

Weight Calculation

```python
#Co-efficient method calling the weight calculated and assign the variable,it is also called slope
weight=regressor.coef_
#used to call the slope value stored in variable.It is the main loop for prediction.
weight
```

Bias Calculation

```python
#Intercept method calling the bias calculated and assign the variable,it is also called origin
bias=regressor.intercept_
#used to call the origin value stored in variable.
bias
```

Evaluvation metrix

```python
# for predict the values pass the input test in predict function and assign variable
Y_pred=regressor.predict(X_test)
```

```python
#import r2 function from sklearn to evaluvate the model
from sklearn.metrics import r2_score
#In evaluvation, function parameters passed as actual output and predict output to have a report for model is good or bad.
r_score=r2_score(Y_test,Y_pred)
#used to call the datas stored in variable(in that the model is best or poor),result observation 0-1.
r_score
```

Save the model

```python
#import the pickle to finalize the good model
import pickle
#create the file name in pickle extension and assign variable
filename="Finalized_model_linear.sav"
```

```python
#Insert the model into the file name using dump function with write binary 
pickle.dump(regressor,open(filename,'wb'))
```

Load the Model

```python
#Load the saved model for checking whether it is working or not and assign variable with read binary
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))
#predict the result of loaded model(verification)
result=loaded_model.predict([[13]])
#used to call the result stored in variable.
result
```
